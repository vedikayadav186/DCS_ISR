```python
# ==============================================
# Color Feature Extraction using OpenCV & Matplotlib
# ==============================================

# ====== Install Required Libraries ======
!pip install opencv-python-headless numpy matplotlib --quiet

# ====== Import Libraries ======
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ====== Step 1: Read the Image ======
#  In Jupyter, use a direct file path instead of files.upload()
# Example path (update this for your image)
filename = "flower.jpg"

# Read image
img = cv2.imread(filename)

if img is None:
    raise FileNotFoundError("Image not found. Check the file path!")

# Convert BGR (OpenCV default) to RGB (for Matplotlib display)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# ====== Step 2: Display the Original Image ======
plt.figure(figsize=(5,5))
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()

# ====== Step 3: Extract Color Histogram Features ======
colors = ('r', 'g', 'b')
plt.figure(figsize=(8,5))
for i, color in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)
    plt.xlim([0, 256])

plt.title("Color Histogram (Feature Representation)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()

# ====== Step 4: Compute Mean RGB Values ======
mean_colors = cv2.mean(img)[:3]
print(f"Mean R: {mean_colors[0]:.2f}, G: {mean_colors[1]:.2f}, B: {mean_colors[2]:.2f}")
```
